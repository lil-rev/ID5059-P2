```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in training data
train = pd.read_csv("C:/Users/ffinl/OneDrive/Documents/Python Scripts/train (1).csv")

# read in test data
test = pd.read_csv("C:/Users/ffinl/OneDrive/Documents/Python Scripts/test.csv")

# Clear the maximum number of columns to be displayed, so that all will be visible.
pd.set_option('display.max_columns', None)

display(train)
display(test)
```

```python
# set id as index
train = train.set_index('id')
test = test.set_index('id')

# split train set into X_train, y_train
y_train = train['Class']
X_train = train.drop('Class', axis = 1)
```

# Solving issuues of unbalanced data

We have an issue of a highly unbalanced dataset. We may be able to resolve this issue using SMOTE (Synthetic Minority Over-sampling Technique).

```python
# lets just check how much of our train set is fraudulent

fraud = train['Class'].sum()

fraud_percent = fraud/len(train['Class'])*100

print(fraud, len(train['Class']))
print(fraud_percent,'%')
```

We are looking at only 0.2% of our training data being of the positive (fraudulent) case. Let's apply SMOTE.

```python
# import SMOTE and RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
# apply SMOTE to oversample the minority class
smote = SMOTE(sampling_strategy='minority')
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

# apply random undersampling to balance the class distribution
rus = RandomUnderSampler(sampling_strategy='majority')
X_train_bal, y_train_bal = rus.fit_resample(X_train_bal, y_train_bal)
```

```python
# lets just check how much of our resampled set is fraudulent

fraud_bal = y_train_bal.sum()

fraud_percent_bal = fraud_bal/len(y_train_bal)*100

print(fraud_bal, len(y_train_bal))
print(fraud_percent_bal,'%')
```

This has rebalanced the train set to 50% fraudulent. This probably far too high due to overlap.

# Fitting to the Logistic Regression

```python
# as this is a "linear" regression, we should center and scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_bal_scaled = scaler.fit_transform(X_train_bal)

X_test_bal_scaled = scaler.transform(test)
```

```python
# fit the logistic regression
from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression(random_state=42)
logistic_mod_bal = logistic_regression_model.fit(X_bal_scaled, y_train_bal)
```

```python
# predict on test data
logistic_predict_bal = logistic_mod_bal.predict(X_test_bal_scaled)

# Create a new DataFrame with 'id' and 'predicted' columns
logistic_results_balanced = pd.DataFrame({'id': test.index, 'Class': logistic_predict_bal})

# display whole set
display(logistic_results_balanced)
# show the fraudulent cases
display(logistic_results_balanced[logistic_results_balanced['Class'] != 0])
```

```python
#probability interpretation
predicted_prob_bal = logistic_mod_bal.predict_proba(X_test_bal_scaled)

# we want the prediction that it is fraudulent (right column)
logistic_fraud_prob_bal = predicted_prob_bal[:,1]

# Create a new DataFrame with 'id' and 'predicted' columns
logistic_prob_bal_results = pd.DataFrame({'id': test.index, 'Class': logistic_fraud_prob_bal})

# display likely fraudulent transactions
display(logistic_prob_bal_results[logistic_prob_bal_results['Class'] > 0.5])
```

```python
# Save the DataFrame as a CSV file
logistic_prob_bal_results.to_csv('logistic_predict_prob_bal.csv', index=False)
```
